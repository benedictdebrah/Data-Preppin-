```python
#importing necessary libraries
import pandas as pd
```

```python
#importing the dataset
df = pd.read_csv('PD 2023 Wk 1 Input.csv')
df.shape
```

```python
df.head()
```

```python
#from the bank column there is whitespace before it ,so i will be removing the spaces 
df.columns = df.columns.str.lstrip()
```

1.Split the transaction code to extract the letters at the start of the the transaction code,this identifies the bank who processes the transaction

```python
#spliting the transaction code
df[['Bank','a','b','c']] = df['Transaction Code'].str.split("-",expand = True)
```

```python
df.head()
```

```python
df.columns
```

```python
#dropping the columns we do not need
df = df[['Bank','Value', 'Customer Code', 'Online or In-Person',
       'Transaction Date']]
```

```python
df.head()
```

```python
df['Online or In-Person'] = df['Online or In-Person'].replace([1,2],['Online', 'In-person'])
```

```python
df.head()
```

```python
df['Online or In-Person'].unique()
```

3. Changing date to the day of the week.

```python
df['Transaction Date'] = pd.to_datetime(df['Transaction Date'])
df['Transaction Date'] = df['Transaction Date'].dt.day_name()
```

```python
df.head()
```

Different levels of detail are required in the outputs. You will need to sum up the values of the transactions in three ways :
1. Total Values of Transactions by each bank
2. Total Values by Bank, Day of the Week and Type of Transaction (Online or In-Person)
3. Total Values by Bank and Customer Code

```python
#Task 1
task1 = df.groupby('Bank')['Value'].sum()
```

```python
task1 = task1.reset_index()
```

```python
task1
```

```python
#Task 2
task2 = df.groupby(['Bank','Transaction Date','Online or In-Person'])['Value'].sum()
```

```python
task2 = task2.reset_index()
task2 = task2[['Bank','Transaction Date','Online or In-Person','Value']]
task2
```

```python
#Task 3
task3 = df.groupby(['Bank','Customer Code'])['Value'].sum()
```

```python

task3 = task3.reset_index()
task3 = task3[['Bank','Customer Code','Value']]
task3
```
