```python
import pandas as pd
import os
```

```python

# Set directory where files are located
directory = r'C:\Users\Benedict\Documents\GitHub\Data preppin\Data-Preppin-\W8'

# Create empty list to hold dataframes
df_list = []

# Loop through all files in directory
for filename in os.listdir(directory):
    if filename.endswith(".csv") and filename.startswith("MOCK"):
        # Read CSV file into a pandas dataframe
        filepath = os.path.join(directory, filename)
        df = pd.read_csv(filepath)
        # Append dataframe to list
        df_list.append(df)

# Concatenate dataframes together
concatenated_df = pd.concat(df_list)
```

```python
concatenated_df = concatenated_df.rename(columns={'Market Cap':'Market Capitalisation'})
```

```python
concatenated_df['Purchase Price'] = concatenated_df['Purchase Price'].str.replace('$', '').astype(float)
```

```python
concatenated_df.dtypes
```

```python

# create the bins and labels for the categories
bins = [0, 25000, 50000, 75000, 100000]
labels = ['Low', 'Medium', 'High', 'Very High']

# use the cut function to create the category column
concatenated_df['Purchase Price Category'] = pd.cut(concatenated_df['Purchase Price'], bins=bins, labels=labels)
```

```python
concatenated_df.head()
```

```python
concatenated_df['Market Capitalisation'] = concatenated_df['Market Capitalisation'].str.replace('$','')
```

```python
# Define a list of market cap values
market_cap = [25000000, 500000000, 75000000000, 150000000000]

# Define a function to convert market cap values to size categories
def convert_market_cap(value):
    if value < 100000000:
        return 'Small'
    elif value < 1000000000:
        return 'Medium'
    elif value < 100000000000:
        return 'Large'
    else:
        return 'Huge'
    
# Create a DataFrame with the market cap values
concatenated_df['market_cap'] = pd.DataFrame({'market_cap': market_cap})

# Create an empty list to store the size categories
size_categories = []

# Loop over the market cap values and apply the convert_market_cap function to each value
for value in market_cap:
    size_category = convert_market_cap(value)
    size_categories.append(size_category)

concatenated_df['size_category'] = concatenated_df['market_cap'].apply(convert_market_cap)
```

```python
# Define a list of column names to keep
cols_to_keep = ['Market Capitalisation', 'size_category', 'Purchase Price Category','Ticker', 'Sector', 'Market', 'Stock Name', 'Market Capitalisation', 'Purchase Price']

# Use .loc to select only the desired columns
df = concatenated_df.loc[:, cols_to_keep]
```
